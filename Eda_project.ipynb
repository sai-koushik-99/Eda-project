```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/vehicle_sales_dirty.csv")
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ["Brand_dup","Selling_Price_dup"],axis = 1, inplace= True)
df.info()
```

```python
df["Brand"] = df["Brand"].ffill()
df["Fuel_Type"] = df["Fuel_Type"].ffill()
df["Transmission"] = df["Transmission"].bfill()
df["Owner"] = df["Owner"].bfill()
df["Selling_Price"] = df["Selling_Price"].fillna(df["Selling_Price"].mode()[0])
df.info()
```

```python
df["Brand"].unique()
```

```python
df["Brand"] = df["Brand"].replace('ford','Ford')
df["Brand"] = df["Brand"].replace('toyota','Toyota')
df['Brand'] = df['Brand'].fillna(df['Brand'].mode()[0])
df["Brand"].unique()
```

```python
df["Fuel_Type"].unique()
```

```python
df["Fuel_Type"] = df["Fuel_Type"].replace("diesel","Diesel")
df["Fuel_Type"] = df["Fuel_Type"].replace("petrol","Petrol")
df["Fuel_Type"] = df["Fuel_Type"].fillna(df['Fuel_Type'].mode()[0])
df["Fuel_Type"].unique()
```

```python
df["Transmission"].unique()
```

```python
df["Transmission"] = df["Transmission"].replace("auto","Automatic")
df["Transmission"] = df["Transmission"].replace("manual","Manual")
df["Transmission"].unique()
```

```python
df["Seller_Type"].unique()
```

```python
df["Seller_Type"] = df["Seller_Type"].replace("dealer","Dealer")
df["Seller_Type"] = df["Seller_Type"].replace("individual","Individual")
df["Seller_Type"].unique()
```

```python
df["Owner"].unique()
```

```python
df["Owner"] = df["Owner"].replace("1st","First")
df["Owner"] = df["Owner"].replace("2nd","Second")
df["Owner"] = df["Owner"].replace("3rd","Third")
df["Owner"] = df["Owner"].fillna(df["Owner"].mode()[0])
df["Owner"].unique()
```

```python
df["Car_ID"].unique()
```

```python
df["Model_Year"].unique()
```

```python
df["Driven_KMs"].unique()
```

```python
df["Engine_CC"].unique()
```

```python
df["Mileage_kmpl"].unique()
```

```python
df["Selling_Price"].unique()
```

```python
df["Present_Price"].unique()
```

```python
clean_df = pd.read_csv("/content/clean_vehicle.csv")
clean_df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = clean_df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,4))
brand_price = df.groupby('Brand')['Selling_Price'].mean().sort_values(ascending=False)
sns.barplot(x=brand_price.index, y=brand_price.values)
plt.xticks(rotation=45)
plt.title("Average Selling Price by Brand")
plt.ylabel("Average Selling Price")
plt.xlabel("Brand")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Present_Price', y='Selling_Price', hue='Fuel_Type', data=df)
plt.title("Selling Price vs Present Price")
plt.xlabel("Present Price ()")
plt.ylabel("Selling Price ()")
plt.show()
```

```python
sns.boxplot(x='Owner', y='Selling_Price', data=df, palette='cool')
```

```python
sns.scatterplot(x='Driven_KMs', y='Selling_Price', hue='Fuel_Type', data=df)
plt.title(" Selling Price vs Driven Kilometers")
plt.show()
```

```python
sns.scatterplot(x='Engine_CC', y='Selling_Price', hue='Transmission', data=df)
plt.title("Selling Price vs Engine CC")
plt.show()
```

```python
sns.scatterplot(x='Mileage_kmpl', y='Selling_Price', hue='Fuel_Type', data=df)
plt.title("Selling Price vs Mileage (kmpl)")
plt.show()
```

```python
sns.boxplot(x='Fuel_Type', y='Selling_Price', data=df, palette='Set2')
plt.title("Selling Price by Fuel Type")
plt.show()
```

```python
sns.boxplot(x='Transmission', y='Selling_Price', data=df, palette='pastel')
plt.title("Selling Price by Transmission")
plt.show()
```

```python
df['Selling_Price'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Selling Price")
plt.xlabel("Selling Price ()")
plt.show()
```

```python
yearly = df.groupby('Model_Year')['Selling_Price'].mean()
plt.figure(figsize=(8,5))
sns.lineplot(x=yearly.index, y=yearly.values, marker='o', color='coral')
plt.title("Average Selling Price by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Average Price ()")
plt.show()
```
